```python
#impute data
def bulk_fill_value(df,column,value="UNDEF"):
    df[column].fillna(value, inplace = True)
    return df
```

```python
train = bulk_fill_value(train,"author","-NO AUTHOR-")
train = bulk_fill_value(train,"title","not provided")
train = bulk_fill_value(train, "text", "NULL")

test = bulk_fill_value(test,"author","-NO AUTHOR-")
test = bulk_fill_value(test,"title","not provided")
test = bulk_fill_value(test, "text", "NULL")
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Maggie Haberman, Matthew Rosenberg and Glenn Thrush")

def get_classification_of_author(df):
    authors = df['author']
    classification = []
    for x in authors:
        if(x == "-NO AUTHOR-"):
            classification.append("NONE")
        else:
            smth = nlp(x)
            for ent in smth.ents:
                if(ent.label_ == "EVENT" or ent.label_ == "ORDINAL"):
                    classification.append("ORG")
                    break
                classification.append(ent.label_)
                break
            if(len(smth.ents) == 0):
                classification.append("ANOTHER")

    df['author_type'] = classification
    return df
```

```python
train = get_classification_of_author(train)
test = get_classification_of_author(test)
```

```python
def get_author_frequency(df):
    authors = df['author']
    visited = []
    frequency = []
    for x in authors:
        count = 0
        if x not in visited:
            for y in authors:
                if(y==x):
                    count+=1
            visited.append(x)
        frequency.append(count)
    df['author_frequency'] = frequency
    return df

def advanced_frequency(df):
    authors = df['author_frequency']
    for x in range(len(authors)):
        if(authors[x]<5):
            if(" and " in df.iloc[x,1]):
                people = df.iloc[x,1].split(" and ")
                count = 0 - len(people)
                for person in people:
                    for var in df['author']:
                        if(person in var):
                            count+=1
                df.loc[x,'author_frequency'] = count
    return df
            
    
```

```python
train = advanced_frequency(get_author_frequency(train))
test = advanced_frequency(get_author_frequency(test))
```

```python
import pickle
import re
from nltk.stem import WordNetLemmatizer

with open("stopwords.txt", "rb") as fp:
    stopwords = set(pickle.load(fp))

def text_cleaning(df, column):
    titles = df[column]
    wordnet = WordNetLemmatizer()
    for i in range(len(titles)):
        title = titles[i]
        title = re.sub(r'\s+', ' ', re.sub('[^A-Za-z0-9]', ' ', title.strip().lower())).strip().split(" ")

        common = stopwords.intersection(set(title))
        for stopword in common:
            try:
                while True:
                    title.remove(stopword)
            except ValueError:
                pass
        
        title = [wordnet.lemmatize(word) for word in title]
        title = ' '.join(title)
        titles[i] = title
    
    df[column] = titles
    return df
```

```python
train = text_cleaning(train,"title")
train = text_cleaning(train,"text")

test = text_cleaning(test,"title")
test = text_cleaning(test,"text")
```

```python
def one_hot_a_feature(df, column, vocab_size = 5000, sent_length = 20):
    values = df[column]
    onehot_repr=[one_hot(words,vocab_size)for words in values]
    embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
    return embedded_docs
```

```python
train_title_one_hot = one_hot_a_feature(train, "title", 5000,20)
train_author_type_one_hot = one_hot_a_feature(train, "author_type",13,1)
train_text_one_hot = one_hot_a_feature(train, "text",5000,500)
train_author_frequency = np.array(train['author_frequency']).reshape(-1,1)

test_title_one_hot = one_hot_a_feature(test, "title", 5000,20)
test_author_type_one_hot = one_hot_a_feature(test, "author_type",13,1)
test_text_one_hot = one_hot_a_feature(test, "text",5000,500)
test_author_frequency = np.array(test['author_frequency']).reshape(-1,1)
```

```python
def compile_data_from_extracted_features(title_one_hot,author_type_one_hot,text_one_hot,author_frequency):
    full_data = np.concatenate((title_one_hot,author_type_one_hot,text_one_hot,author_frequency), axis=1)
    return full_data
```

```python
X = compile_data_from_extracted_features(train_title_one_hot,train_author_type_one_hot,train_text_one_hot,train_author_frequency)
X_test = compile_data_from_extracted_features(test_title_one_hot,test_author_type_one_hot,test_text_one_hot,test_author_frequency)
y = np.array(train['label']).reshape(-1,1)
```

```python
np.savetxt('X_train.txt',X, fmt='%d')
np.savetxt('X_test.txt',X_test, fmt='%d')
np.savetxt('y.txt',y, fmt='%d')
```
